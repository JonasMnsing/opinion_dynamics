```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
p_vals      = [0.05,0.1,0.5,0.9]
avg_opinion = [np.load(f"../data/avg_opinion_p={p}.npy") for p in p_vals]
int_density = [np.load(f"../data/int_density_p={p}.npy") for p in p_vals]
con_matrix  = [np.load(f"../data/connection_matrix_p={p}.npy") for p in p_vals]
```

```python
fig = plt.figure()
ax  = fig.add_subplot()
for i, avg_op in enumerate(avg_opinion):
    ax.plot(avg_op, label=f'p = {p_vals[i]}')
ax.set_xlabel("Iteration")
ax.set_ylabel("Average Opinion")
ax.legend()
```

```python
fig = plt.figure()
ax  = fig.add_subplot()
for i, int_dens in enumerate(int_density):
    ax.plot(int_dens, label=f'p = {p_vals[i]}')
ax.set_xlabel("Iteration")
ax.set_ylabel("Interface Density")
ax.legend()
```

```python
i   = 0
G   = nx.from_numpy_array(con_matrix[i])
# pos = nx.shell_layout(G)
pos = nx.kamada_kawai_layout(G)

nx.draw(G, pos=pos, node_size=50, )
```
